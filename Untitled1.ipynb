```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import pi
import cv2
import scipy.misc
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
from imgaug import augmenters as iaa
```

```python
# Hyperparameters
split_ratio = 0.8 
# split_ratio = 0.7
# split_ratio = 0.9 

optimizer_learning_rate = 10**-4
# optimizer_learning_rate = 10**-3

#drop_val = 0.5
drop_val = 0.8
```

```python
directory = "D:\Windsor\DL for Vision & NLP\Project-2\driving_dataset"
data_file = os.path.join(directory, "data.txt")

img_paths = []
y = []

with open(data_file) as f:
    for line in f:
        image_name, angle = line.split()
        
        img_path = os.path.join(directory, image_name)
        img_paths.append(img_path)
        
        angle_radians = float(angle) * (pi / 180) 
        y.append(angle_radians)
y = np.array(y)
```

```python
plt.figure()
img = cv2.imread(img_paths[100])
plt.imshow(img)
plt.show()
```

```python
split = int(len(img_paths) * split_ratio)

x_train = img_paths[:split]
y_train = y[:split]

x_test = img_paths[split:]
y_test = y[split:]

print("Number of Training Samples:" + str(len(x_train)))
print("Number of Testing Samples: " + str(len(x_test)))
```

```python
fig = plt.figure(figsize = (10, 7))
plt.hist(y_train, bins = 50, histtype = "step")
plt.hist(y_test, bins = 50, histtype = "step")
plt.title("Steering Wheel angle in train and test")
plt.xlabel("Angle")
plt.ylabel("Bin count")
plt.show()
```

```python
def preprocess(img):
    img = img[-150:]
    #img = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
    #img = cv2.GaussianBlur(img,  (3, 3), 0)
    img = cv2.resize(img, (200, 66))
    img = img/255.0
    return img
```

```python
original_image = cv2.imread(x_train[50])
img = preprocess(original_image)
fig, axs = plt.subplots(1, 2, figsize=(15, 10))
fig.tight_layout()
axs[0].imshow(original_image)
axs[0].set_title('Original Image')
axs[1].imshow(img)
axs[1].set_title('Preprocessed Image')
```

```python
def random_augment(image, steering_angle):
    if np.random.rand() < 0.5:
        image = zoom(image)
    if np.random.rand() < 0.5:
        image = img_random_brightness(image)
    if np.random.rand() < 0.5:
        image, steering_angle = img_random_flip(image, steering_angle)
    return image, steering_angle

def zoom(img):
    zoom = iaa.Affine(scale=(1, 1.3))
    img = zoom.augment_image(img)
    return img

def img_random_brightness(image):
    brightness = iaa.Multiply((0.2, 1.2))
    image = brightness.augment_image(image)
    return image

def img_random_flip(image, steering_angle):
    image = cv2.flip(image,1)
    steering_angle = -steering_angle
    return image, steering_angle
```

```python
def loadBatch(batch_size, istraining, batch):
    x_batch = []
    y_batch = []
    for i in range(batch_size):
        if istraining:
            #random_index = random.randint(0, len(train_x) - 1)
            #img = cv2.imread(train_x[random_index])
            img = cv2.imread(x_train[(batch + i)  % len(x_train)])
            # angle = y_train[random_index]
            angle = y_train[(batch + i)  % len(x_train)]
            img, angle = random_augment(img, angle)
        else:
            #random_index = random.randint(0, len(test_x) - 1)
            #img = cv2.imread(x_test[random_index])
            #angle = y_test[random_index]
            img = cv2.imread(x_test[(batch + i)  % len(x_test)])
            angle = y_test[(batch + i)  % len(x_test)]
        
    img = preprocess(img)
    x_batch.append(img)
    y_batch.append(angle)
    return np.asarray(x_batch), np.asarray(y_batch)
```

```python
x_train_batch, y_train_batch = loadBatch(1, 1, 0)
x_test_batch, y_test_batch = loadBatch(1, 0, 0)
 
fig, axs = plt.subplots(1, 2, figsize=(15, 10))
fig.tight_layout()
 
axs[0].imshow(x_train_batch[0])
axs[0].set_title('Training Image')

axs[1].imshow(x_test_batch[0])
axs[1].set_title('Testing Image')
```

```python
def weight(shp):
    initial = tf.truncated_normal(shape = shp, stddev = 0.1)
    return tf.Variable(initial) 

def bias(shp):
    initial = tf.constant(0.1, shape=shp)
    return tf.Variable(initial)

x_input = tf.placeholder(tf.float32, shape = [None, 66, 200, 3],           )
y_true = tf.placeholder(tf.float32, name = "Plc_2")
dropout = tf.placeholder(tf.float32)

#-------------------------- First Convolutional Layer --------------------------
W1 = weight([5,5,3,24])
B1 = bias([24])
C1 = tf.nn.relu(tf.nn.conv2d(x_input, W1, strides = [1, 2, 2, 1], padding = "VALID") + B1)
#-------------------------- Second Convolutional Layer -------------------------
W2 = weight([5,5,24,36])
B2 = bias([36])
C2 = tf.nn.relu(tf.nn.conv2d(C1, W2, strides = [1, 2, 2, 1], padding = "VALID") + B2)
#-------------------------- Third Convolutional Layer --------------------------
W3 = weight([5,5,36,48])
B3 = bias([48])
C3 = tf.nn.relu(tf.nn.conv2d(C2, W3, strides = [1, 2, 2, 1], padding = "VALID") + B3)
#-------------------------- Fourth Convolutional Layer -------------------------
W4 = weight([3,3,48,64])
B4 = bias([64])
C4 = tf.nn.relu(tf.nn.conv2d(C3, W4, strides = [1, 1, 1, 1], padding = "VALID") + B4)
#-------------------------- Fifth Convolutional Layer --------------------------
W5 = weight([3,3,64,64])
B5 = bias([64])
C5 = tf.nn.relu(tf.nn.conv2d(C4, W5, strides = [1, 1, 1, 1], padding = "VALID") + B5)


#---------------------- First Fully Connected Dense Layer ----------------------
W_FC1 = weight([1152, 1164])
B_FC1 = bias([1164])
FC1 = tf.nn.relu(tf.matmul(tf.reshape(C5, [-1, 1152]), W_FC1) + B_FC1)
FC1_dropout = tf.nn.dropout(FC1, dropout)

#---------------------- Second Fully Connected Dense Layer ---------------------
W_FC2 = weight([1164, 100])
B_FC2 = bias([100])
FC2 = tf.nn.relu(tf.matmul(FC1_dropout, W_FC2) + B_FC2) 
FC2_dropout = tf.nn.dropout(FC2, dropout)

#---------------------- Third Fully Connected Dense Layer ----------------------
W_FC3 = weight([100, 50])
B_FC3 = bias([50])
FC3 = tf.nn.relu(tf.matmul(FC2_dropout, W_FC3) + B_FC3)
FC3_dropout = tf.nn.dropout(FC3, dropout)

#---------------------- Fourth Fully Connected Dense Layer ---------------------
W_FC4 = weight([50, 10])
B_FC4 = bias([10])
FC4 = tf.nn.relu(tf.matmul(FC3_dropout, W_FC4) + B_FC4)
FC4_dropout = tf.nn.dropout(FC4, dropout)

#--------------------- Final Layer with Tangent Function -----------------------
W_FC5 = weight([10, 1])
B_FC5 = bias([1])
y_predicted = tf.identity(tf.matmul(FC4_dropout, W_FC5) + B_FC5)
```

```python
model_dir = "./Models/"
session = tf.InteractiveSession()

L2NormConst = 0.001
#L2NormConst = 0.0001
train_vars = tf.trainable_variables() 

# MSE Loss
loss = tf.reduce_mean(tf.square(tf.subtract(y_true, y_predicted))) + tf.add_n([tf.nn.l2_loss(w) for w in train_vars]) * L2NormConst

train_step = tf.train.AdamOptimizer(learning_rate = optimizer_learning_rate).minimize(loss)

session.run(tf.global_variables_initializer())
saver = tf.train.Saver()

no_of_epochs = 15
batch = 0
batch_size = 100
epoch_arr = []
train_loss_arr = []
test_loss_arr = []

for epoch in range(no_of_epochs):
    avg_train_loss = 0
    avg_test_loss = 0
    te_loss_old = 5000
    
    for i in range(int(len(img_paths)/batch_size)):
        train_batch_x, train_batch_y = loadBatch(batch_size, 1 , batch)
        train_step.run(feed_dict = {x_input: train_batch_x, y_true: train_batch_y, dropout: drop_val})
        tr_loss = loss.eval(feed_dict = {x_input: train_batch_x, y_true: train_batch_y, dropout: 1.0})
        avg_train_loss = avg_train_loss + (tr_loss / batch_size)

        test_batch_x, test_batch_y = loadBatch(batch_size, 0, batch)
        te_loss_new = loss.eval(feed_dict = {x_input: test_batch_x, y_true: test_batch_y, dropout: 1.0})
        avg_test_loss = avg_test_loss + (te_loss_new / batch_size)

        print("Epoch {} - Train Loss: {}, Test Loss: {} ".format(epoch+1, tr_loss, te_loss_new))

        te_loss_old = te_loss_new
        batch = batch + batch_size
            
    if not os.path.exists(model_dir):
        os.makedirs(model_dir)
    path = os.path.join(model_dir, "model.ckpt")
    saver.save(sess = session, save_path = path)
    print("Succesfully saved Model at {} after {} epoch.".format(path, epoch + 1))        
    epoch_arr.append(epoch)
    train_loss_arr.append(avg_train_loss)
    test_loss_arr.append(avg_test_loss)
    
df = pd.DataFrame(columns = ["Epoch", "Train Loss", "Test Loss"])
df["Epoch"] = epoch_arr
df["Train Loss"] = train_loss_arr
df["Test Loss"] = test_loss_arr
df.to_csv(os.path.join(model_dir, "log.csv"), index = False)
```
